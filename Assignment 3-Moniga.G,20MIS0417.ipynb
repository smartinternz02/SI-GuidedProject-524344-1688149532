## 5. Handle the Missing values.

```python
valid_columns = data.select_dtypes(include=[np.number]).columns

# Fill missing values with mean of valid columns
data[valid_columns] = data[valid_columns].fillna(data[valid_columns].mean())
data
```

## 7. Check for Categorical columns and perform encoding.


#### Generally categorical column involves the column that are not numerical and involve some categorical division instead of the numerical values and numerical manipulation 

```python
data.info()#has the data type as value so it is easier to find which are not numerical and add them as the numerical data.
```

```python
#we have to focus on the column which are described as objects
#Label Encoder-they generally assign the numerical value for each category
#One hot handling -splits them to the number of classes
from sklearn.preprocessing import LabelEncoder
categorical_columns = data.select_dtypes(include=['object']).columns#converting the objects as they are categorical columns

# Perform encoding for each categorical column
for column in categorical_columns:
    
    label_encoder = LabelEncoder()
    data[column] = label_encoder.fit_transform(data[column])
print(data)#all the categorical columns are converted to numerical columns
```

## 11. Build the Model

```python
#building different regression models
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression()
from sklearn.tree import DecisionTreeClassifier
df=DecisionTreeClassifier(criterion='entropy',random_state=0)
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

# 12. Train the Model

```python
lor.fit(X_train,y_train)
df.fit(X_train,y_train)
knn.fit(X_train,y_train)
lr.fit(X_train,y_train)
```

# 13. Test the Model


```python
y_pred1 = lr.predict(X_test)

y_pred2 = lor.predict(X_test)

y_pred3 = knn.predict(X_test)

y_pred4 = df.predict(X_test)
```

## 14. Measure the performance using Metrics.

```python
from sklearn.metrics import mean_squared_error,r2_score, mean_absolute_error
```

```python
#for linear regression
mse1=mean_squared_error(y_test,y_pred1)
mae1=mean_absolute_error(y_test,y_pred1)
rmse1=np.sqrt(mse1)
r2lr=r2_score(y_test,y_pred1)


print("Mean Absolute Error for linear regression(MAE):", mae1)
print("Mean Squared Error for linear regression (MSE):", mse1)
print("Root Mean Squared Error for linear regression(RMSE):", rmse1)
print("R-squared (R2)for linear regression:", r2lr)
```

```python
#for logistic regression
mse2=mean_squared_error(y_test,y_pred2)
mae2=mean_absolute_error(y_test,y_pred2)
rmse2=np.sqrt(mse2)
r2lor=r2_score(y_test,y_pred2)


print("Mean Absolute Error for logistic regression(MAE):", mae2)
print("Mean Squared Error for logistic regression (MSE):", mse2)
print("Root Mean Squared Error for logistic regression(RMSE):", rmse2)
print("R-squared (R2)for logistic regression:", r2lor)
```

```python
#for KNN regression
mse3=mean_squared_error(y_test,y_pred3)
mae3=mean_absolute_error(y_test,y_pred3)
rmse3=np.sqrt(mse3)
r2knn=r2_score(y_test,y_pred3)


print("Mean Absolute Error for KNN regression(MAE):", mae3)
print("Mean Squared Error for KNN regression (MSE):", mse3)
print("Root Mean Squared Error for KNN regression(RMSE):", rmse3)
print("R-squared (R2)for KNN regression:", r2knn)
```

```python
#for decision tree
mse4=mean_squared_error(y_test,y_pred4)
mae4=mean_absolute_error(y_test,y_pred4)
rmse4=np.sqrt(mse4)
r2dt=r2_score(y_test,y_pred4)


print("Mean Absolute Error for decision tree(MAE):", mae4)
print("Mean Squared Error for decision tree (MSE):", mse4)
print("Root Mean Squared Error for decision tree(RMSE):", rmse4)
print("R-squared (R2)for decision tree:", r2dt)
```
