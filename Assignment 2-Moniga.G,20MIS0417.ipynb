## 5. Handle the Missing values.


```python
#if we have null value in the numerical column then we have to replace them with Mean or Median
# else if the column that has the null value is categorical based then it should be replaced with Mode
data.mean()
```

```python
data.median()
```

```python
data.mode()
```

```python
#let us check for null values in the data set
data.isna()
```

```python
data.isnull().any()
```

```python
data.isnull().sum()
```

```python
#it seems like we have null value for age so we have to replace them with the mean/median as they are numerical value
data['age'].fillna(data['age'].mean(),inplace=True)
```

```python
data.isnull().sum()# now we have filled the null value of the age
```

```python
#similarly the deck,embark_town and embarked has some null value and these are categorical column
#embarked
mode_value = data['embarked'].mode()[0]
data['embarked'].fillna(mode_value,inplace=True)#have to use mode
```

```python
data.isnull().sum()# now we have filled the null value of the embarked 
```

```python
#deck
mode_value = data['deck'].mode()[0]
data['deck'].fillna(mode_value,inplace=True)#have to use mode
data.isnull().sum()# now we have filled the null value of the deck
```

```python
#embark_town
mode_value = data['embark_town'].mode()[0]
data['embark_town'].fillna(mode_value,inplace=True)#have to use mode
data.isnull().sum()# now we have filled the null value of the embark_town 0 indicates the absence of the null value
```

#### missing values are handled
